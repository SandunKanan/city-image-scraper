```python
import json

# Load the image seeds
with open('./JSON/image_seeds.json', 'r') as file:
    image_seeds = json.load(file)

# Load the city image URLs
with open('./JSON/city_image_urls.json', 'r') as file:
    city_image_urls = json.load(file)

# Create a list of city names in the order of image_seeds
city_names = [city for city in city_image_urls.keys()]

# Update the URLs in image_seeds
for i, seed in enumerate(image_seeds):
    city_name = city_names[i]
    if city_name in city_image_urls:
        new_url = city_image_urls[city_name]
        seed['mob_url'] = new_url
        seed['web_url'] = new_url

# Save the updated image seeds back to the file
with open('./JSON/updated_image_seeds.json', 'w') as file:
    json.dump(image_seeds, file, indent=2)

print("Updated image_seeds.json with new URLs and saved as updated_image_seeds.json")
```
